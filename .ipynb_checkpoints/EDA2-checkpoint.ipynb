Let's split our sequences so that we take the last elements of each sequence and put them into a list to be used as our targets. The remaining elements in each sequence will be used as the set of observations.

```python
def window_transform_series(sequences):
    """
    Split sequences into Observations and Label lists. Extract and reove the last term of each sequence and put them 
    in a separate ordered list.

    :param x: List of sequences.
    :return: X and y list. Where X contains the obserations and y de labels.
    """
    X = []
    y = []
    
    for seq in sequences:
        X_seq, y_seq = split_sequence(seq)
        X.extend(X_seq)
        y.extend(y_seq)
        
#         X.append(np.asarray(seq[0: -1]))
#         y.append(seq[-1])

    # reshape each
    X = np.asarray(X)
    y = np.asarray(y)

    return X, y

def split_sequence(seqeunce):
    
    X = []
    y = []
    
    inverted = np.array(seqeunce)[::-1]
    for idx, token in enumerate(inverted[:-1]):
        X.append(np.asarray(inverted[idx+1:][::-1]))
        y.append(token)
        
    X = np.asarray(X)
    y = np.asarray(y)

    return X, y
        
    
    
```

Now we need to pad our sequences so they are all the same length. Let's pad them to the maximum length of the sequences in set X.

So the maximum length of all sequences in set X is 10. This makes sense since we had realized that the overal maximum length was 11. After removing the last entry it makes sense now that the max length is 10.

Lets pad our sequences so that they are all the same length. We will fill them up with zeros until they are the size of the maximum length of all the sequences.

```python
def pad_sequece(x, length=None):
    """
    Pad x
    :param x: List of sequences.
    :param length: Length to pad the sequence to.  If None, use length of longest sequence in x.
    :return: Padded numpy array of sequences
    """
    return pad_sequences(x,maxlen=length,padding='post')
```

```python
def one_hot_encode_matrix(sequences, dimension):
    processed_seqs = []
    for seq in sequences:
        processed_seqs.append(one_hot_encode_list(seq, dimension))
    return processed_seqs

def one_hot_encode_list(seq, dimension):
    encoded_sequences = []
    for token in seq:
        encoded = [0]*dimension
        encoded[token-1] = 1
        encoded_sequences.append(encoded)
        
    return encoded_sequences
    
```

```python
# Plot training & validation loss values
def plot_model_loss(model_history):
    plt.figure(figsize=(15,7));
    plt.plot(model_history.history['loss'])
    plt.plot(model_history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    
plot_model_loss(model_history)
```

```python
def predict_next_cities(model,input_cities,num_to_predict): 
    
    predicted_cities = input_cities
    for i in range(num_to_predict):
        
        print('Input sequence: ' + str(input_cities))
        padded = pad_sequece([input_cities], length = max_seq_length)
        print('Padded  sequence: ' + str(padded))
        # Hot encode the padded input sequence
        X_test = one_hot_encode_matrix(padded, len(city_list))
        # Reshape it to be put into the model
        X_test = np.asarray(np.reshape(X_test, (len(X_test), max_seq_length, len(city_list))))

        # Make this round's prediction
        test_predict = model.predict(X_test,verbose = 0)[0]
        
        # Get the 5 most relevant predictions (the classes with probabilities closest to one)
        highest_predictions = np.argpartition(test_predict, -5)[-5:]
        print('Indexes of predictions with highest scores: ' + str(highest_predictions))
        # Get the probability scores for each one of the most relevant predictions
        scores_preds = test_predict[highest_predictions]
        print('Probability distributions: ' + str(scores_preds))
        # Sort the predicted values by the most relevant first (the ones with probability closest to the leftmost side of the list)
        # Note that we are adding + 1 to realign the predicted value with the city token (class)
        ordered_predictions = highest_predictions[np.argsort(test_predict[highest_predictions])][::-1] + 1
        print('Reindexed sorted predictions: ' + str(ordered_predictions))
        print('----')
        
        for pred in ordered_predictions:
            # Check if the predicted city is already part of the list and skip it if so
            if pred in predicted_cities:
                continue
            else:
                predicted_cities.append(pred)
                break
                
    return(predicted_cities)
```

Lets see how our model does on the set that we reserved for testing.

```python
def decode_vectors(binary_matrix):
    tokens = []
    for vect in binary_matrix:
        tokens.append(np.argmax(vect)+1)
    return tokens;


y_tokens = decode_vectors(y_test)
# print(len(y_tokens))
print(y_tokens[:11])

```

```python
def evaluate_model(model, test_data, prediction_interval):
    test_predict = model.predict(test_data,verbose = 1)

    score = 0
    for idx, pred in enumerate(test_predict):
        # Get the most relevant predictions (the classes with probabilities closest to one)
        highest_predictions = np.argpartition(pred, -prediction_interval)[-prediction_interval:]
        print('Predictions with highest scores: ' + str(highest_predictions))
        # Get the probability scores for each one of the most relevant predictions
        scores_preds = test_predict[highest_predictions]
    #     print('Probability distributions: ' + str(scores_preds))
        # Sort the predicted values by the most relevant first (the ones with probability closest to the leftmost side of the list)
        ordered_predictions = highest_predictions[np.argsort(pred[highest_predictions])][::-1] + 1
        print('Sorted predictions: ' + str(ordered_predictions))

        expected_y = y_tokens[idx]
        index_of_expected = np.where(ordered_predictions == expected_y)[0] 

        if index_of_expected:
            score += 1
            print('Found in first: ' + str(prediction_interval))
        else:
            print('Not found in first: ' + str(prediction_interval))
        print('----')
        
    return score
    

```
